# Car Price Dataset Cleaning and Preprocessing

```python
import pandas as pd

import warnings

warnings.filterwarnings("ignore")
```

# Step 1: Load the dataset

```python
df = pd.read_csv("C:/Users/ibiib/Project_Sem1/Data Source/car_prices.csv")
```

# Step 2: Check the dataset

```python
print(df.head())
print(df.info())
print(df.isnull().sum())
```

# Step 3: Remove duplicate rows

```python
df = df.drop_duplicates()
```

# Step 4: Fill missing values in numeric columns with the median

```python
for col in df.select_dtypes(include=['number']).columns:
    df[col] = df[col].fillna(df[col].median())
```

# Step 5: Fill missing values in categorical columns with the mode

```python
for col in df.select_dtypes(include=['object']).columns:
    mode = df[col].mode(dropna=True)

    if len(mode) > 0:
        df[col] = df[col].fillna(mode.iloc[0])
    else:
        df[col] = df[col].fillna("Unknown")
```

# Step 6: Remove leading and trailing spaces from text columns

```python
for col in df.select_dtypes(include=['object']).columns:
    df[col] = df[col].astype(str).str.strip()
```

# Step 7: Verify missing values are handled

```python
print(df.isnull().sum())
```
