# Import Libraries

```python
import pandas as pd        # This is our main tool for handling tables/spreadsheets
import numpy as np         # Great for math operations — we'll use this a lot
import matplotlib.pyplot as plt  # This draws our charts and graphs
import seaborn as sns      # Makes prettier charts than plain matplotlib
import warnings

# This just hides annoying warning messages that don't affect our work
warnings.filterwarnings('ignore')
```

# Load Dataset

```python
# Load the CSV file into a DataFrame

df = pd.read_csv('C:/Users/ibiib/Project_Sem1/Data Source/car_prices.csv')

print(f"Rows: {df.shape[0]:,} | Columns: {df.shape[1]}")

df.head()
```

# Drop Unnecessary Columns

```python
# 'vin' is a unique vehicle ID
# It has no analytical value for price analysis, so we drop it.

df.drop(columns=['vin'], inplace=True)

print(f"Columns remaining: {df.shape[1]}")
print(df.columns)
```

# Missing Values

```python
missing = pd.DataFrame({
    'Missing Count': df.isnull().sum(),
    'Missing %': (df.isnull().sum() / len(df)) * 100
})

missing = missing[missing['Missing Count'] > 0].sort_values('Missing %', ascending=False)
print(missing)
```

# Duplicate Rows

```python
duplicates = df.duplicated().sum()
print(f"Duplicate rows: {duplicates}")
```

# Data Types 

```python
print(df.dtypes)
```

# Basic Statistics

```python
df.describe()
```

# Drop Rows Where Critical Columns Are Missing

```python
# Drop rows where make, model, odometer, mmr, sellingprice or saledate are missing
# These columns are too important to guess or fill

critical_cols = ['make', 'model', 'odometer', 'mmr', 'sellingprice', 'saledate']
df.dropna(subset=critical_cols, inplace=True)

print(f"Rows after dropping critical nulls: {len(df):,}")
```

# Fill Remaining Missing Values

```python
# Fill categorical columns with 'unknown'
categorical_fills = ['transmission', 'body', 'trim', 'color', 'interior']
df[categorical_fills] = df[categorical_fills].fillna('unknown')

# Fill condition with median (more robust than mean for skewed data)
df['condition'].fillna(df['condition'].median(), inplace=True)

print("Missing values remaining:")
print(df.isnull().sum()[df.isnull().sum() > 0])
print("None — all clean!" if df.isnull().sum().sum() == 0 else "")
```

# Remove Suspicious Values

```python
# Remove unrealistic odometer readings
df = df[df['odometer'] <= 300000]

# Remove unrealistic prices
df = df[df['sellingprice'] >= 100]
df = df[df['mmr'] >= 100]

print(f"Rows after removing suspicious values: {len(df):,}")
```

# Fix Data Types + Extract Date Features

```python
# Convert condition to integer — it represents a score, not a decimal
df['condition'] = df['condition'].astype(int)

# Convert saledate from text to proper datetime format
df['saledate'] = pd.to_datetime(df['saledate'], utc=True)

# Extract useful time features while we're here
df['sale_month'] = df['saledate'].dt.month
df['sale_year'] = df['saledate'].dt.year

print(df[['saledate', 'sale_month', 'sale_year']].head())
print()
print(df.dtypes)
```

# Final Cleaned Dataset Check

```python
print(f"Original rows : 558,837")
print(f"Cleaned rows  : {len(df):,}")
print(f"Rows removed  : {558837 - len(df):,}")
print()
print("Missing values remaining:", df.isnull().sum().sum())
df.describe()
```

# Check Sale Year vs Manufacturing Year

```python
# A car cannot be sold in a year earlier than it was manufactured
invalid_dates = df[df['sale_year'] < df['year']]

print(f"Rows where sale year < manufacturing year: {len(invalid_dates):,}")
print()
print(invalid_dates[['year', 'sale_year', 'saledate']].head(10))
```

# Remove Chronologically Invalid Rows

```python
# Keep only rows where the car was sold in or after its manufacturing year
df = df[df['sale_year'] >= df['year']]

print(f"Rows after chronology fix: {len(df):,}")
```

# Check ["body"] inconsistencies

```python
# Print all unique body types to spot inconsistencies
print(df['body'].value_counts().to_string())
```

# Check ["make"] inconsistencies

```python
# Print all unique makes to spot inconsistencies
print(df['make'].value_counts().to_string())
```

# Fix Body Column

```python
# Step 1: Lowercase everything first, then title case to standardize
df['body'] = df['body'].str.lower().str.title()

# Step 2: Fix specific values that title case doesn't handle correctly
body_mapping = {
    'Suv'                    : 'SUV',
    'G Sedan'                : 'Sedan',
    'G Coupe'                : 'Coupe',
    'G Convertible'          : 'Convertible',
    'G37 Coupe'              : 'Coupe',
    'G37 Convertible'        : 'Convertible',
    'Supercrew'              : 'SuperCrew',
    'Supercab'               : 'SuperCab',
    'Crewmax Cab'            : 'CrewMax Cab',
    'Cts Coupe'              : 'CTS Coupe',
    'Cts-V Coupe'            : 'CTS-V Coupe',
    'Cts Wagon'              : 'CTS Wagon',
    'Cts-V Wagon'            : 'CTS-V Wagon',
    'Tsx Sport Wagon'        : 'TSX Sport Wagon',
    'Genesis Coupe'          : 'Coupe',
    'Granturismo Convertible': 'Convertible',
    'Elantra Coupe'          : 'Coupe',
    'Koup'                   : 'Coupe',
    'Q60 Coupe'              : 'Coupe',
    'Q60 Convertible'        : 'Convertible',
    'Beetle Convertible'     : 'Convertible',
    'Regular-Cab'            : 'Regular Cab',
    'E-Series Van'           : 'Van',
    'Promaster Cargo Van'    : 'Van',
    'Transit Van'            : 'Van',
    'Ram Van'                : 'Van',
}

df['body'] = df['body'].replace(body_mapping)

print(f"Unique body types after cleaning: {df['body'].nunique()}")
print(df['body'].value_counts())
```

# Fix Make Column

```python
# Step 1: Lowercase then title case
df['make'] = df['make'].str.lower().str.title()

# Step 2: Fix abbreviations, typos and variants
make_mapping = {
    'Bmw'          : 'BMW',
    'Gmc'          : 'GMC',
    'Gmc Truck'    : 'GMC',
    'Mini'         : 'MINI',
    'Fiat'         : 'FIAT',
    'Hummer'       : 'HUMMER',
    'Vw'           : 'Volkswagen',
    'Volkswagen'   : 'Volkswagen',
    'Landrover'    : 'Land Rover',
    'Land Rover'   : 'Land Rover',
    'Mercedes'     : 'Mercedes-Benz',
    'Mercedes-B'   : 'Mercedes-Benz',
    'Ford Truck'   : 'Ford',
    'Ford Tk'      : 'Ford',
    'Chev Truck'   : 'Chevrolet',
    'Dodge Tk'     : 'Dodge',
    'Mazda Tk'     : 'Mazda',
    'Hyundai Tk'   : 'Hyundai',
}

df['make'] = df['make'].replace(make_mapping)

# Step 3: Remove garbage values that are not real car brands
garbage_makes = ['Dot', 'Airstream']
df = df[~df['make'].isin(garbage_makes)]

print(f"Unique makes after cleaning: {df['make'].nunique()}")
print(df['make'].value_counts())
```

# Save Cleaned Dataset

```python
# Save the cleaned dataframe so we don't need to re-run all cleaning steps
df.to_csv('car_prices_cleaned.csv', index=False)

print(f"Cleaned dataset saved successfully.")
print(f"Shape: {df.shape[0]:,} rows × {df.shape[1]} columns")
```

# Summary Statistics

```python
# Summary statistics for all numerical columns
numerical_cols = ['year', 'condition', 'odometer', 'mmr', 'sellingprice']
df[numerical_cols].describe().T
```

# Distributions (Histograms)

```python
fig, axes = plt.subplots(2, 3, figsize=(15, 8))
fig.suptitle('Distribution of Numerical Variables', fontsize=16)

for i, col in enumerate(numerical_cols):
    ax = axes[i//3, i%3]
    df[col].hist(bins=50, ax=ax, color='steelblue', edgecolor='white')
    ax.set_title(col)
    ax.set_xlabel(col)
    ax.set_ylabel('Frequency')

axes[1, 2].set_visible(False)  # Hide empty subplot
plt.tight_layout()
plt.show()
```

# Boxplots (Outlier Detection)

```python
fig, axes = plt.subplots(1, 5, figsize=(18, 5))
fig.suptitle('Boxplots — Spotting Outliers', fontsize=16)

for i, col in enumerate(numerical_cols):
    axes[i].boxplot(df[col].dropna(), patch_artist=True,
                    boxprops=dict(facecolor='steelblue', color='navy'))
    axes[i].set_title(col)
    axes[i].set_xlabel(col)

plt.tight_layout()
plt.show()
```

# Correlation Matrix

```python
# Calculate correlation between all numerical columns
# Correlation ranges from -1 to 1
# 1 = perfect positive relationship, -1 = perfect negative, 0 = no relationship
numerical_cols = ['year', 'condition', 'odometer', 'mmr', 'sellingprice']
corr_matrix = df[numerical_cols].corr()
print(corr_matrix)
```

# Correlation Heatmap

```python
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm',
            linewidths=0.5, square=True)
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.show()
```
